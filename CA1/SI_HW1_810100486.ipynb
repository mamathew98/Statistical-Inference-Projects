## سوال اول 
در آمار، یک confounder (همچنین متغیر confounding، عامل confounding) متغیری است که هم متغیر وابسته و هم متغیر مستقل را تحت تأثیر قرار می دهد و باعث ایجاد ارتباط جعلی می شود. confounding یک مفهوم علّی است، و به این ترتیب، نمی توان آن را بر حسب correlation یا association توصیف کرد.</br>
متغیر confounding، یک متغیر «اضافی» است که شما آن را در نظر نگرفته اید. این متغیرها می توانند یک آزمایش را خراب کنند و نتایج بی فایده ای به ما بدهند. همچنین می توانند پیشنهاد کنند که میان دو آزمایش همبستگی وجود دارد در حالی که در واقع وجود ندارد. این متغیرها حتی می توانند در نتایج آزمایش bias ایجاد کنند.

## سوال دوم
در اپیدمیولوژی، نرخ مرگ و میر موردی (CFR) - یا گاهی اوقات به طور دقیق‌تر خطر مرگ و میر موردی - نسبت افرادی است که با بیماری خاصی تشخیص داده می‌شوند و در نهایت بر اثر آن می‌میرند. برخلاف میزان مرگ و میر یک بیماری، CFR دوره زمانی بین شروع بیماری و مرگ را در نظر نمی گیرد. یک CFR به طور کلی به صورت درصد بیان می شود. این معیاری از کشندگی بیماری است و ممکن است با درمان های مختلف تغییر کند. CFR ها اغلب برای دوره های گسسته و با زمان محدود، مانند عفونت های حاد استفاده می شوند.

<img src="q2.png" width="700" >

مقصود سوال در حقیقت این است که اگر ما یک رویداد (event) را به ازای تمامی حالات یک متغیر تصادفی بررسی کنیم و نقیض آن رویداد در تمام این حالات احتمال شرطیش بیشتر باشه آیا در نهایت می‌شود گفت احتمال غیرشرطی نقیض به طور کلی بیشتر هست یا خیر؟ </br>
داریم که : </br>
$$
 E(X) = \sum_{x}^{} x*P(X=x) 
$$

امید ریاضی (به انگلیسی: Expected value) یا مقدار چشمداشتی یا مقدار انتظاری در نظریه احتمالات، مقدارِ قابل انتظاری است از یک متغیر تصادفیِ گسسته که برابر است با مجموع حاصل‌ضرب احتمالِ وقوع هر یک از حالات ممکن در مقدار آن حالت. در نتیجه میانگین برابر است با مقداری که به‌طور متوسط از یک فرایند تصادفی با بی‌نهایت تکرار انتظار می‌رود. به بیان ساده‌تر، امید ریاضی از یک متغیر تصادفی، مقدارِ میانگینِ تعداد دفعاتِ مشاهده شدهٔ یک وضعیت است.</br>
با توجه به این تعریف و اینکه مقادیر متغیر تصادفی در دو نمونه با یکدیگر برابر است و به ازای هر مقدار متغیر، احتمال (تعداد وقوع) نمونه چین بیشتر است میتوان گفت انتظار می‌رود که در حالت کلی نیز نمونه‌های چین تعداد بیشتری خواهند داشت و از آنجا که در متن سوال از (should) استفاده شده است و به طور قطع بیان نشده و معنای انتظار داشتن میدهد میتوان این عبارت را قبول کرد. البته این در صورتی است که از بازه زمانی متفاوت که این دو نمونه گرفته شده است صرف نظر شود.

## سوال چهارم
تعمیم پذیری می‌توان به‌عنوان بسط یافته‌ها و نتیجه‌گیری‌های پژوهشی از یک مطالعه انجام شده بر روی یک جامعه نمونه به جامعه کلی تعریف کرد. در حالی که قابلیت اطمینان این عمل مطلق نیست، اما از نظر آماری محتمل است. از آنجا که تعمیم پذیری به داده هایی در مورد جمعیت های بزرگ نیاز دارد، تحقیقات کمی - برای مثال تجربی - بهترین پایه را برای تولید قابلیت تعمیم گسترده فراهم می کند.هر چه جامعه نمونه بزرگتر باشد، می توان نتایج را بیشتر تعمیم داد.

از بسیاری جهات، تعمیم پذیری چیزی بیش از پیش بینی بر اساس یک تجربه تکراری نیست. اگر اتفاقی به طور مکرر رخ دهد، انتظار داریم که در آینده نیز به همین منوال ادامه دهد. محققان هنگام تعمیم یافته های مطالعات خود از همین نوع استدلال استفاده می کنند. هنگامی که محققان داده های کافی برای حمایت از یک فرضیه را جمع آوری کردند، می توان پیش فرضی در رابطه با رفتار آن داده ها فرموله کرده و آن را به شرایط مشابه قابل تعمیم بسازند. با این حال، به دلیل مبنای احتمالی، چنین تعمیم نمی تواند به عنوان قطعی یا جامع در نظر گرفته شود.

در حالی که تعمیم‌پذیری می‌تواند در محیط‌های غیررسمی و غیر آکادمیک رخ دهد، اما معمولاً فقط در مورد روش‌های تحقیق خاصی در مطالعات دانشگاهی اعمال می‌شود. روش های کمی امکان تعمیم پذیری را فراهم می کند. به عنوان مثال، تحقیقات تجربی اغلب نتایج قابل تعمیم ایجاد می کند. با این حال، برای یافتن نتایج قابل تعمیم، چنین آزمایش‌هایی باید دقیق باشد.

سه نوع تعمیم پذیری وجود دارد که برای تولید مدل های احتمالی در تعامل هستند. همه آنها شامل تعمیم یک درمان یا اندازه گیری به جمعیتی خارج از مطالعه اصلی است.

- در یک نوع تعمیم پذیری، محققان تعیین می کنند که آیا یک درمان خاص نتایج یکسانی را در شرایط مختلف ایجاد می کند یا خیر. برای انجام این کار، آنها باید تصمیم بگیرند که آیا یک جنبه در محیط اصلی، عاملی فراتر از درمان، نتیجه خاصی را ایجاد کرده است یا خیر. این نشان می دهد که چگونه درمان با شرایط جدید سازگار است

- شکل دوم تعمیم پذیری به جای درمان، بر اندازه گیری ها متمرکز است. برای اینکه یک نتیجه خارج از گروه آزمایشی قابل تعمیم در نظر گرفته شود، باید نتایج یکسانی را با اشکال مختلف اندازه گیری ایجاد کند.

- نوع سوم تعمیم پذیری مربوط به آزمودنی های وضعیت است. اگرچه نتایج یک آزمایش ممکن است از لحاظ داخلی معتبر باشد، یعنی برای گروه آزمایش شده قابل اجرا باشد، اما در بسیاری از شرایط نمی توان نتایج را فراتر از آن گروه خاص تعمیم داد.

مهم نیست که این سه شکل تعمیم‌پذیری چقدر با دقت اعمال می‌شوند، تضمین مطلقی وجود ندارد که نتایج به‌دست‌آمده در یک مطالعه در هر موقعیتی خارج از مطالعه رخ دهد. به منظور تعیین روابط علی در یک محیط آزمایشی، دقت از اهمیت بالایی برخوردار است. با این حال، اگر محققان بخواهند یافته های خود را تعمیم دهند، دامنه و واریانس باید بیش از دقت مورد تاکید قرار گیرد. بنابراین، آزمایش دقیق و تعمیم پذیری به طور همزمان دشوار می شود، زیرا تمرکز بر یکی باعث کاهش قابلیت اطمینان دیگری می شود. یک راه حل برای این مشکل گرفتن تعداد بیشتری از مشاهدات است که اثر دوگانه دارد: اول، جمعیت نمونه را افزایش می دهد که تعمیم پذیری را افزایش می دهد. دوم، دقت را می توان به طور منطقی حفظ کرد، زیرا خطاهای تصادفی بین مشاهدات به طور میانگین محاسبه می شود


علیت حوزه‌ای از آمار است که معمولاً توسط افراد به اشتباه درک می‌شود. این باور اشتباه وجود دارد که که چون داده‌ها با یکدیگر همبستگی نشان می‌دهند، لزوماً یک رابطه علّی بین آنها وجود دارد.ارتباط یا همبستگی بین متغیرها به سادگی نشان می دهد که مقادیر با هم تغییر میکنند. لزوماً نشان نمی دهد که تغییرات در یک متغیر باعث تغییر در متغیر دیگر می شود. اثبات علیت می تواند دشوار باشد.

استفاده از مطالعه کنترل شده مؤثرترین راه برای ایجاد علیت بین متغیرها است. در یک مطالعه کنترل شده، نمونه یا جامعه به دو بخش تقسیم می شود که هر دو گروه تقریباً از هر نظر قابل مقایسه هستند. سپس دو گروه درمان های متفاوتی دریافت می کنند و نتایج هر گروه ارزیابی می شود.

یک دلیل این اتفاق رایج یک متغیر Confounding است که یک همبستگی جعلی ایجاد می کند. یک متغیر مداخله گر با هر دو متغیر مورد بررسی همبستگی دارد. این امکان وجود دارد که متغیر Confounding باعث ایجاد این سو برداشت شود!

نقاط ضعف نمودار دایره‌ای
- مقادیر دقیق داده قابل شناسایی نیست

```r
library(dplyr)

# Data transformation
pi_df <- mtcars %>% 
  group_by(cyl) %>% # Variable to be transformed
  count() %>% 
  ungroup() %>% 
  mutate(perc = `n` / sum(`n`)) %>% 
  arrange(perc) %>%
  mutate(labels = scales::percent(perc)) %>%
  mutate(groups = sprintf("%d Cylinder",cyl) )
pi_df
```

```r
ggplot(pi_df, aes(x = "", y = perc, fill = groups)) +
  geom_col() +
  geom_text(aes(label = labels),
            position = position_stack(vjust = 0.5)) +
  coord_polar(theta = "y")+
  theme_void()
```
