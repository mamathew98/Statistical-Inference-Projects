```r
library("data.table")
```

```r
pres_conf <- fread("alldecisiondata_withrt_long.csv", header = TRUE, sep = ",",select = c("Pressure", "conf"))
pres_conf
```

```r
pressure = pres_conf[pres_conf$Pressure == "Pressure", ]
pressure
```

```r
control = pres_conf[pres_conf$Pressure == "Control", ]
control
```

```r
res = t.test(control[[2]], pressure[[2]])
res
```

```r
uncert_conf <- fread("alldecisiondata_withrt_long.csv", header = TRUE, sep = ",",select = c("uncertainty", "conf"))
uncert_conf
```

```r
ou = uncert_conf[uncert_conf$uncertainty == "OU", ]
ou
```

```r
mu = uncert_conf[uncert_conf$uncertainty == "MU", ]
mu
```

```r
res = t.test(ou[[2]], mu[[2]])
res
```
